## Plot activations

```python
from matplotlib import pyplot as plt 
import numpy as np
```

```python
def sigmoid(x):
    return 1 / (1 + np.e ** -x)
def tanh(x):
    return np.tanh(x)
def relu(x):
    return np.array ([i if i >= 0 else 0.0 for i in x])
def leaky_relu(x,k):
    return np.array ([i if i >= 0 else k*i for i in x])
```

```python
s_x = np.linspace(-5,2,100)
plt.plot(s_x,sigmoid(s_x),color='r')
plt.plot(s_x,tanh(s_x),color='g')
plt.plot(s_x,relu(s_x),color='b')
plt.plot(s_x,leaky_relu(s_x,.1),color='y')
plt.legend(['sigmoid', 'tanh','relu','leaky_relu'], loc='upper left')
plt.show()
```

```python
def dsig(x):
    h = 1e-5
    return (sigmoid(x+h)-sigmoid(x-h))/(2*h)
def dtan(x):
    h = 1e-5
    return (tanh(x+h)-tanh(x-h))/(2*h)
def drel(x):
    h = 1e-5
    return (relu(x+h)-relu(x-h))/(2*h)
def dlrel(x,k):
    h = 1e-5
    return (leaky_relu(x+h,k)-leaky_relu(x-h,k))/(2*h)
s_x = np.linspace(-5,5,100)
plt.plot(s_x,dsig(s_x),color='r')
plt.plot(s_x,dtan(s_x),color='g')
plt.plot(s_x,drel(s_x),color='b')
plt.plot(s_x,dlrel(s_x,.1),color='y')
plt.legend(['sigmoid', 'tanh','relu','leaky_relu'], loc='upper left')
plt.show()
```
