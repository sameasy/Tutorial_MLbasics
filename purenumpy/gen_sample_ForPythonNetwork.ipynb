## This code generates training and test distributions and samples

```python
# libraries
import math
import matplotlib.pyplot as plt
import numpy as np
sample_size = 2000
# create data
x1 = np.random.normal(loc=7.5,scale=2,size=sample_size)
y1 = np.random.normal(loc=7.5,scale=2,size=sample_size)
l1 = np.ones(sample_size)
h1 = np.zeros(sample_size)
s1 = np.dstack((x1,y1,l1,h1))
s1 = s1.reshape(sample_size,4)
np.savetxt("sample1.txt", s1)
#x1a = np.array(x1)

#x2 = np.random.normal(loc=20,scale=2,size=sample_size)
#y2 = np.random.normal(loc=10,scale=2,size=sample_size)
x2 = np.array([])
y2 = np.array([])
for i in range(5,15,1):
#    print i
    tmpx = np.random.normal(loc=i,scale=1,size=sample_size/10)
    x2 = np.hstack((x2,tmpx))
    tmpy = np.random.normal(loc=math.sqrt(15**2-i**2),scale=1,size=sample_size/10)
    y2 = np.hstack((y2,tmpy))


l2 = np.zeros(sample_size)
h2 = np.ones(sample_size)
s2 = np.dstack((x2,y2,l2,h2))
s2 = s2.reshape(sample_size,4)
np.savetxt("sample2.txt", s2)

comb = np.concatenate((s1, s2), axis=0)
print comb
np.random.shuffle(comb)
#print comb
train_frac = 80./100.*2*sample_size
print train_frac
np.savetxt("sample_shuff_train.txt", comb[:int(train_frac)])
np.savetxt("sample_shuff_test.txt", comb[int(train_frac):])
#np.savetxt("sample1.txt", (a,b,1), fmt="%d")
#print x1a
```
